```python
import openmc
import numpy
```

```python
#define materials

#u02
uranium = openmc.Material(name='uranium')
uranium.add_nuclide('U235', 0.03)
uranium.add_nuclide('U238', 0.97)
uranium.add_nuclide('O16', 2)
uranium.set_density('g/cm3', 10)

#zirconium
zirconium = openmc.Material(name='zirconium')
zirconium.add_nuclide('Zr', 1)
zirconium.set_density('g/cm3', 6.6)

#water
water = openmc.Material(name='water')
water.add_nuclide('H1', 2)
water.add_nuclide('O16', 1)
water.set_density('g/cm3', 1)
water.add_s_alpha_beta('c_H_in_H2O')

#creazione materiali
materials = openmc.Materials([uranium, zirconium, water])
materials.export_to_xml()
```
